# Random Forest Classification

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
RF=RandomForestClassifier(n_estimators=20,random_state=10)
```

```python
clf=RF.fit(iris.data,iris.target)
```

```python
clf.estimators_   # Can get information of all Decision Tree Results
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf.estimators_[2],filled=True)#max_depth
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf.estimators_[19],filled=True)
```
