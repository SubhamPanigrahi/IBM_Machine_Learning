```python
# Importing Libraries
import numpy as np
import pandas as pd
```

```python
# Importing the Car Sales data 
df = pd.read_csv("Data/Car_sales.csv")
```

```python
# checking Null values
df.isnull().sum()
```

```python
df.head()
```

```python
# Changing the Numeric type NaN to mean.
for x in df.columns:
    if(x == 'Model' or x == 'Manufacturer' or  x == 'Vehicle type' or  x == 'Latest Launch' or  x == 'Unnamed: 15'):
        continue
    df[x].fillna(value = np.mean(df[x]), inplace = True)
df
```

```python
df['Manufacturer'].mode()
```

```python
# Treating the Non-NUMERIC data
df['Manufacturer'].fillna(value = 'Dodge', inplace = True)
df
```

```python
df.isnull().sum()
```

```python
# Dropping the Unnamed Column.
df.drop(['Unnamed: 15'], axis =  1)
```

```python
# No. of Rows and Columns
print('Number of Columns = ', len(df.columns))
print('Number of Records = ', len(df.index))
```

```python
# Min Sales
df[(df['Sales in thousands'] == df['Sales in thousands'].min())][['Manufacturer','Model','Sales in thousands']]
```

```python
# MAx Sales
df[(df['Sales in thousands'] == df['Sales in thousands'].max())][['Manufacturer','Model','Sales in thousands']]
```

```python
# Lowest passenger sales
df[(df['Sales in thousands'] == df['Sales in thousands'].min()) & (df['Vehicle type'] == 'Passenger')][['Manufacturer','Model','Sales in thousands']]
```

```python
# Highest sales with engine size> 6
df[(df['Engine size'] > 6)][['Manufacturer','Model','Sales in thousands']]
```

```python
# Highest fuel efficiency as a car
df1 = df[df['Vehicle type'] == 'Car']
df1[df1['Fuel efficiency'] == df1['Fuel efficiency'].max()][['Manufacturer','Fuel efficiency']]
```

```python
# company name who has got lowest mileage with respect to Vehicle Type as Passengers.
df1 = df[df['Vehicle type'] == 'Passenger']
df1[df1['Fuel efficiency'] == df1['Fuel efficiency'].min()][['Manufacturer','Fuel efficiency']]
```

```python
# company name who has got least price and maximum number of sales figures.
df['MaxSalesProfit'] = df['Sales in thousands']/df['Price in thousands']
df[df['MaxSalesProfit'] == df['MaxSalesProfit'].max()][['Manufacturer','MaxSalesProfit']]
```

```python
# company name who has got highest price and minimum number of sales figures.
df[df['MaxSalesProfit'] == df['MaxSalesProfit'].max()][['Manufacturer','MaxSalesProfit']]
```

```python
# the model who has got maximum number of Horsepower
df[df['Horsepower'] == max(df['Horsepower'])][['Model','Horsepower']]
```

```python
# model and company name who has got maximum mileage with lowest sales figures
df[df['Fuel efficiency'] == max(df['Fuel efficiency'])][['Model', 'Manufacturer']]
```

```python
# model and company name who has got minimum mileage with highest sales figures.
df1 = df[df['MaxSalesProfit']==max(df['MaxSalesProfit'])]
df1[df1['Fuel efficiency'] == min(df1['Fuel efficiency'])][['Model','Manufacturer']]
```

```python
# model and company name where the wheel base is maximum with Vehicle type as Car
df1 = df[df['Vehicle type'] == 'Car']
df1[df1['Wheelbase']==max(df1['Wheelbase'])][['Model','Manufacturer']]
```

```python
# model and company name where the wheel base is minimum with Vehicle type as Passengers
df1 = df[df['Vehicle type'] == 'Passenger']
df1[df1['Wheelbase']==min(df1['Wheelbase'])][['Model','Manufacturer']]
```

```python
# model and company name where the curb weight is maximum with maximum mileage
df1 = df[df['Curb weight']==max(df['Curb weight'])]
df1[df1['Fuel efficiency'] == max(df1['Fuel efficiency'])][['Model','Manufacturer']]
```

```python
# model and company name where the curb weight is minimum with maximum mileage
df1 = df[df['Curb weight']==max(df['Curb weight'])]
df1[df1['Fuel efficiency'] == min(df1['Fuel efficiency'])][['Model','Manufacturer']]
```

```python
# model and company name where the curb weight is minimum with minimum mileage
df1 = df[df['Curb weight']==min(df['Curb weight'])]
df1[df1['Fuel efficiency'] == min(df1['Fuel efficiency'])][['Model','Manufacturer']]
```

```python
# Change the name of column x_4_year to 4_Year_Resale_Value
df.rename(columns = {'x_4_year' : '4_Year_Resale_Value'}, inplace = True)
df.head()
```

```python
# model and company name whose sales and Resale Value is Highest
df['q18'] = df['4_Year_Resale_Value']*df['Sales in thousands']
ans = df[df['q18'] == max(df['q18'])][['Model','Manufacturer']]
df.drop(columns = 'q18', inplace = True)
ans
```

```python
# model and company name whose sales, Mileage and Resale Value is Highest but Price is lowest
df['q19'] = (df['4_Year_Resale_Value']*df['Sales in thousands']*df['Fuel efficiency'])/df['Price in thousands']
ans = df[df['q19'] == max(df['q19'])][['Model','Manufacturer']]
df.drop(columns = 'q19', inplace = True)
ans
```

```python
# model and company name whole resale value is lowest but sale is highest.
df['q'] = df['4_Year_Resale_Value']/df['Sales in thousands']
ans = df[df['q'] == max(df['q'])][['Model','Manufacturer']]
df.drop(columns = 'q', inplace = True)
ans
```

```python
# Replace the observations in Sales colum that are greater than the mean of Sales column
df1 = df[df['Sales in thousands'] > np.mean(df['Sales in thousands'])]
to_put = np.mean(df1['Sales in thousands'])
df.loc[df['Sales in thousands'] > np.mean(df['Sales in thousands']), 'Sales in thousands'] = to_put
df.head()
```

```python
# Replace the observations in Price column that are greater than the mean of Price column.
df1 = df[df['Price in thousands'] > np.mean(df['Price in thousands'])]
to_put = np.mean(df1['Price in thousands'])
df.loc[df['Price in thousands'] > np.mean(df['Price in thousands']), 'Price in thousands'] = to_put
df.head()
```

```python
# Replace the observations in Horsepower that are greater than the mean of Horsepower column.
df1 = df[df['Horsepower'] > np.mean(df['Horsepower'])]
to_put = np.mean(df1['Horsepower'])
df.loc[df['Horsepower'] > np.mean(df['Horsepower']), 'Horsepower'] = to_put
df.head()
```
