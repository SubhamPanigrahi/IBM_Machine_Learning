```python
# data who have got Highest Salary with no working experience with ssc_p > 66
df1 = df[(df['workex']=='No') & (df['ssc_p'] > 66)]
df1[df1['salary'] == max(df1['salary'])]
```

```python
# Replace the observations in salary that are greater than the mean of salary column.
# (Replacing the salaries of all those records which are greater than the mean salary with their mean so thata the mean won't change)
print('Mean Salary before replacement = ', np.mean(df['salary']))
df1 = df[df['salary'] > np.mean(df['salary'])]
replacement_mean = np.mean(df1['salary'])
df.loc[df.salary > np.mean(df['salary']) , 'salary'] = replacement_mean
print('Mean Salary after replacement = ', np.mean(df['salary']))
df
```

```python
# Replace the observations in ssc_p that are greater than the mean of ssc_p
df1 = df[df['ssc_p'] > np.mean(df['ssc_p'])]
replacement_mean = np.mean(df1['ssc_p'])
print(replacement_mean)
df.loc[df.ssc_p > np.mean(df['ssc_p']) , 'ssc_p'] = replacement_mean
df
```

```python
# Repalce the observations in hsc_p that are greater than the mean of hsc_p
df1 = df[df['hsc_p'] > np.mean(df['hsc_p'])]
replacement_mean = np.mean(df1['hsc_p'])
print(replacement_mean)
df.loc[df.hsc_p > np.mean(df['hsc_p']) , 'hsc_p'] = replacement_mean
df
```

```python
# Replace the observations in degree_p that are greater than the mean of degree_p
df1 = df[df['degree_p'] > np.mean(df['degree_p'])]
replacement_mean = np.mean(df1['degree_p'])
print('Replacement = ' , replacement_mean)
df.loc[df.degree_p > np.mean(df['degree_p']) , 'degree_p'] = replacement_mean
df
```

```python
# Replace the observations in etest_p that are greater than the mean of etest_p
df1 = df[df['etest_p'] > np.mean(df['etest_p'])]
replacement_mean = np.mean(df1['etest_p'])
print('Replacement = ' , replacement_mean)
df.loc[df.etest_p > np.mean(df['etest_p']) , 'etest_p'] = replacement_mean
df
```

```python
# Replace the observations in specialisation from Marketing & HR  to HR
df.loc[df.specialisation == 'Mkt&HR', 'specialisation'] = 'HR'
df
```

```python
# Replace the observations in ssc_b from Others to PYTHON
df.loc[df.ssc_b == 'Others' , 'ssc_b'] = 'Python'
df
```

```python
# Replace the observations in degree_t from Sci&Tech to Sci&Com
df.loc[df.degree_t == 'Sci&Tech' , 'degree_t'] = 'Sci&Com'
df
```
